```python
!pip install pandas datetime matplotlib tqdm
import pandas
from datetime import datetime
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
datelist = pandas.date_range(start=datetime(2018, 1, 23), end=datetime(2020, 1, 19)).tolist()
```

```python
data = []

#format YYYYMMDD
target_url = 'http://content.caiso.com/green/renewrpt/{}_DailyRenewablesWatch.txt'
for date in tqdm(datelist):
    try:
        url = target_url.format(date.strftime('%Y%m%d'))

        renewables = pandas.read_csv(url, sep='\t', nrows=24, skiprows=2, header=None)
        renewables.dropna(inplace=True, axis=1)
        renewables.columns = ['hour', 'geothermal', 'biomass', 'biogas', 'small hydro', 'wind', 'solar', 'solar thermal']


        resources = pandas.read_csv(url, sep='\t', nrows=24, skiprows=30, header=None)
        resources.dropna(inplace=True, axis=1)
        resources.columns = ['hour', 'renewables', 'nuclear', 'thermal', 'imports', 'hydro']


    #     plt.plot(renewables['wind'])
    #     plt.plot(renewables['solar'])
    #     plt.plot(renewables['wind'] + renewables['solar'])

    #     plt.plot(resources['thermal'])
    #     plt.legend(['wind', 'solar', 'CEF', 'thermal'])
    #     plt.title(date.date())
    #     plt.show()

        sub_frame = pandas.DataFrame([renewables['wind'], renewables['solar'], resources['thermal']])
        sub_frame = sub_frame.T
        sub_frame['date'] = date
        sub_frame['date'] = pandas.to_datetime(sub_frame['date'])
        sub_frame['hour'] = sub_frame.index
        sub_frame = sub_frame.set_index(['date', 'hour'])
        data.append(sub_frame)
    except Exception as e:
        print(url)
        print(e)
    
```
